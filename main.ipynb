```python
%matplotlib inline

import numpy as np
import tensorflow as tf
import os

from PIL import Image
from matplotlib import pyplot as plt
```

```python
# Rotation by degree/axes

generator = tf.keras.preprocessing.image.ImageDataGenerator(
    rotation_range=90
)
```

```python
# Image import
image = '/home/probe/imageManIp/asset/image/train/media.jpeg'
plt.imshow(plt.imread(image));
```

```python
x, y = next(generator.flow_from_directory('asset', batch_size=1))
plt.imshow(x[0].astype('uint8'))    # rotation variance plot on every iteration
```

```python
# Dimensions shift
# Just for lookups - https://keras.io/api/preprocessing/image/
generator = tf.keras.preprocessing.image.ImageDataGenerator(
    width_shift_range=[-200, -100, 0, 50, 100],
    height_shift_range=[-100, 0, 100]
)
```

```python
x, y = next(generator.flow_from_directory('asset', batch_size=1))
plt.imshow(x[0].astype('uint8'))    # shift variance plot on every iteration
```

```python
# Exposure compensation

generator = tf.keras.preprocessing.image.ImageDataGenerator(
    brightness_range=[0, 5]     # list for offset or tuples
)
```

```python
x, y = next(generator.flow_from_directory('asset', batch_size=1))
plt.imshow(x[0].astype('uint8'))    # brightness variance plot on every iteration
```
